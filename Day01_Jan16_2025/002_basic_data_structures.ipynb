```python
class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            raise IndexError("pop from empty stack")

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            raise IndexError("peek from empty stack")

    def size(self):
        return len(self.items)

    def clear(self):
        self.items = []

    def __str__(self):
        return str(self.items)

# Example usage
stack = Stack()
print("Initial stack:", stack)

stack.push(1)
stack.push(2)
stack.push(3)
print("Stack after pushing 1, 2, 3:", stack)

print("Peek top element:", stack.peek())

print("Pop top element:", stack.pop())
print("Stack after popping:", stack)

print("Is stack empty?", stack.is_empty())

print("Stack size:", stack.size())

stack.clear()
print("Stack after clearing:", stack)
print("Is stack empty after clearing?", stack.is_empty())
```

```python
class Queue:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)
        else:
            raise IndexError("dequeue from empty queue")

    def peek(self):
        if not self.is_empty():
            return self.items[0]
        else:
            raise IndexError("peek from empty queue")

    def size(self):
        return len(self.items)

    def clear(self):
        self.items = []

    def __str__(self):
        return str(self.items)

# Example usage
queue = Queue()
print("Initial queue:", queue)

queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
print("Queue after enqueuing 1, 2, 3:", queue)

print("Peek front element:", queue.peek())

print("Dequeue front element:", queue.dequeue())
print("Queue after dequeuing:", queue)

print("Is queue empty?", queue.is_empty())

print("Queue size:", queue.size())

queue.clear()
print("Queue after clearing:", queue)
print("Is queue empty after clearing?", queue.is_empty())
```

```python
from collections import deque

# List
my_list = [1, 2, 3, 4, 5]
print("List:", my_list)
my_list.append(6)
print("List after append:", my_list)
my_list.remove(3)
print("List after remove:", my_list)
print("List length:", len(my_list))
print("List slice [1:3]:", my_list[1:3])

# Tuple
my_tuple = (1, 2, 3, 4, 5)
print("Tuple:", my_tuple)
print("Tuple length:", len(my_tuple))
print("Tuple slice [1:3]:", my_tuple[1:3])

# Set
my_set = {1, 2, 3, 4, 5}
print("Set:", my_set)
my_set.add(6)
print("Set after add:", my_set)
my_set.remove(3)
print("Set after remove:", my_set)
print("Set length:", len(my_set))
print("Set union with {4, 5, 6}:", my_set.union({4, 5, 6}))

# Dictionary
my_dict = {'a': 1, 'b': 2, 'c': 3}
print("Dictionary:", my_dict)
my_dict['d'] = 4
print("Dictionary after adding key 'd':", my_dict)
del my_dict['b']
print("Dictionary after deleting key 'b':", my_dict)
print("Dictionary keys:", my_dict.keys())
print("Dictionary values:", my_dict.values())

# Deque
my_deque = deque([1, 2, 3, 4, 5])
print("Deque:", my_deque)
my_deque.append(6)
print("Deque after append:", my_deque)
my_deque.appendleft(0)
print("Deque after appendleft:", my_deque)
my_deque.pop()
print("Deque after pop:", my_deque)
my_deque.popleft()
print("Deque after popleft:", my_deque)
print("Deque length:", len(my_deque))
```
