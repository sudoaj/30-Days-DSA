```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.head.next = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.head

    def display(self):
        nodes = []
        temp = self.head
        if self.head is not None:
            while True:
                nodes.append(temp.data)
                temp = temp.next
                if temp == self.head:
                    break
        print(" -> ".join(map(str, nodes)))

# Example usage:
cll = CircularLinkedList()
cll.append(1)
cll.append(2)
cll.append(3)
cll.display()
```

```python
# Append more elements to the circular linked list
cll.append(4)
cll.append(5)

# Display the updated circular linked list
cll.display()
```

```python
class CircularLinkedList:
    def __init__(self):
        self.head = None
    
    def display(self):
        nodes = []
        temp = self.head
        if self.head is not None:
            while True:
                nodes.append(temp.data)
                temp = temp.next
                if temp == self.head:
                    break
        print(" -> ".join(map(str, nodes)))
    
    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.head.next = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.head

    def prepend(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.head.next = self.head
        else:
            new_node.next = self.head
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
            self.head = new_node

    def delete(self, key):
        if self.head:
            if self.head.data == key:
                if self.head.next == self.head:
                    self.head = None
                else:
                    temp = self.head
                    while temp.next != self.head:
                        temp = temp.next
                    temp.next = self.head.next
                    self.head = self.head.next
            else:
                temp = self.head
                prev = None
                while temp.next != self.head:
                    prev = temp
                    temp = temp.next
                    if temp.data == key:
                        prev.next = temp.next
                        temp = temp.next

    def search(self, key):
        temp = self.head
        if self.head:
            while True:
                if temp.data == key:
                    return True
                temp = temp.next
                if temp == self.head:
                    break
        return False
cll = CircularLinkedList()
# Example usage:
cll.prepend(0)
cll.display()
cll.delete(3)
cll.display()
print(cll.search(2))
print(cll.search(3))
```

```python
import time

class Process:
    def __init__(self, name, burst_time):
        self.name = name
        self.burst_time = burst_time

class RoundRobinScheduler:
    def __init__(self, time_quantum):
        self.time_quantum = time_quantum
        self.processes = CircularLinkedList()

    def add_process(self, name, burst_time):
        process = Process(name, burst_time)
        self.processes.append(process)

    def schedule(self):
        current = self.processes.head
        while current:
            process = current.data
            if process.burst_time > 0:
                print(f"Processing {process.name} for {min(self.time_quantum, process.burst_time)} units")
                time.sleep(1)  # Simulate the time taken for processing
                process.burst_time -= self.time_quantum
                if process.burst_time <= 0:
                    print(f"{process.name} has finished execution")
            current = current.next
            if current == self.processes.head:
                all_done = True
                temp = self.processes.head
                while True:
                    if temp.data.burst_time > 0:
                        all_done = False
                        break
                    temp = temp.next
                    if temp == self.processes.head:
                        break
                if all_done:
                    break

# Example usage:
scheduler = RoundRobinScheduler(time_quantum=2)
scheduler.add_process("P1", 5)
scheduler.add_process("P2", 3)
scheduler.add_process("P3", 7)
scheduler.schedule()
```

```python
class Song:
    def __init__(self, title, duration):
        self.title = title
        self.duration = duration

class Playlist:
    def __init__(self):
        self.songs = CircularLinkedList()

    def add_song(self, title, duration):
        song = Song(title, duration)
        self.songs.append(song)

    def remove_song(self, title):
        temp = self.songs.head
        if self.songs.head:
            while True:
                if temp.data.title == title:
                    self.songs.delete(temp.data)
                    break
                temp = temp.next
                if temp == self.songs.head:
                    break

    def display_playlist(self):
        temp = self.songs.head
        if self.songs.head:
            while True:
                print(f"Title: {temp.data.title}, Duration: {temp.data.duration} mins")
                temp = temp.next
                if temp == self.songs.head:
                    break

    def play(self):
        temp = self.songs.head
        if self.songs.head:
            while True:
                print(f"Playing {temp.data.title} for {temp.data.duration} mins")
                time.sleep(temp.data.duration)  # Simulate playing the song
                temp = temp.next
                if temp == self.songs.head:
                    break

# Example usage:
playlist = Playlist()
playlist.add_song("Song 1", 3)
playlist.add_song("Song 2", 4)
playlist.add_song("Song 3", 5)
playlist.display_playlist()
playlist.play()
playlist.remove_song("Song 2")
playlist.display_playlist()
```

```python
# Add more songs to the playlist
playlist.add_song("Song 4", 2)
playlist.add_song("Song 5", 3)

# Display the updated playlist
playlist.display_playlist()

# Play the songs in the updated playlist
playlist.play()
```
